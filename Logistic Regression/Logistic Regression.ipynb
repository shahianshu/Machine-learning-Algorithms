```python
#logistic Regression function

def hypothesis(x,theta):
    h=np.dot(x,theta)
    return sigmoid(h)

def sigmoid(z):
    return 1.0/(1.0+np.exp(-1.0*z))


def error(y_true,x,theta):
    m=x.shape[0]
    err=0.0
    for i in range(m):
        hx=hypothesis(x[i],theta)
        err+=y_true[i]*np.log2(hx) +(1-y_true[i])*np.log2(1-hx)
    return -err/m

def get_grad(y_true,x,theta):
    grad=np.zeros(theta.shape)
    m=x.shape[0]
    for i in range(m):
        hx=hypothesis(x[i],theta)
        grad += (y_true[i] - hx)*x[i]
        
    grad/=m
    return grad
        
#for 1 iteration
def grad_asscent(y_true,x,theta,learning_rate):
    
    err=error(y_true,x,theta)
    print(err)
    
    grad=get_grad(y_true,x,theta)
    theta=theta+learning_rate*grad
    
    return err,theta
    


```
