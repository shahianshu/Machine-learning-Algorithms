```python
x,y=make_blobs(centers=5,n_features=2,n_samples=500,random_state=3)
```

```python
plt.figure(0)
plt.grid(True)
plt.scatter(x[:,0],x[:,1],c=y)
plt.show()
```

```python
k=5
colors=["green","red","blue","yellow","orange"]
clusters={}

for i in range(k):
    center=10*(2*np.random.random((x.shape[1],))-1)
    points=[]
    
    cluster={
        'center':center,
        'color':colors[i],
        'points':points
    }
    clusters[i]=cluster
```

```python
clusters
```

```python
def distance(v1,v2):
    return np.sqrt(np.sum((v1-v2)**2))
```

```python
def assignPointsToClusters(clusters):
    for xi in range(x.shape[0]):
        dist=[]
        curr_x=x[xi]
        for ki in range(k):
            d=distance(clusters[ki]['center'],curr_x)
            dist.append(d)
            
        curr_cluster=np.argmin(dist)
        clusters[curr_cluster]['points'].append(curr_x)

        
def updateClusters(clusters):
    for kx in range(k):
        pts = np.array(clusters[kx]['points'])
        if pts.shape[0]>0:
            new_center=pts.mean(axis=0)
            clusters[kx]['center']=new_center
            clusters[kx]['points']=[]

def plotClusters(clusters):
    
    for kx in range(k):
        pts = np.array(clusters[kx]['points'])
#         print(pts.shape)
        
        try:
            plt.scatter(pts[:,0],pts[:,1],color=clusters[kx]['color'])
        except:
            pass
        uk=clusters[kx]['center']
        plt.scatter(uk[0],uk[1],marker='*',color='black')
       
```

```python
assignPointsToClusters(clusters)
plotClusters(clusters)
updateClusters(clusters)
```
