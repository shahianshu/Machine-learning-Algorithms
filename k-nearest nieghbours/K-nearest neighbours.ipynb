```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use('seaborn-pastel')
```

```python
dfx=pd.read_csv('../Datasets/xdata.csv')
dfy=pd.read_csv('../Datasets/ydata.csv')
```

```python
x=dfx.values
y=dfy.values
x=x[:,1:]
y=y[:,1:].reshape((-1,))
print(x.shape,y.shape)
print(y)
```

```python
plt.scatter(x[:,0],x[:,1],c=y)
plt.show()
```

```python
query_x=np.array([2,3])
plt.scatter(x[:,0],x[:,1],c=y)
plt.scatter(query_x[0],query_x[1],color='red')
plt.show()
```

```python
def dist(x1,x2):
    return np.sqrt(sum((x1-x2)**2))

def knn(x,y,query_point,k):
    vals=[]
    
    for i in range(x.shape[0]):
        d=dist(query_point,x[i])
        vals.append((d,y[i]))
        
    
    vals = sorted(vals)
    vals = vals[:k]
    vals=np.array(vals)
    
    new_vals = np.unique(vals[:,1],return_counts=True)
    index=new_vals[1].argmax() #gives the index of the element having max count
    prediction=new_vals[0][index]
    return prediction
    
```

```python
ans=knn(x,y,query_x,5)
print(ans)
```
